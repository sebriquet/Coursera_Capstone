# Exploring Neighborhoods in Madrid 

### Background & Problem

Madrid has one of the largest metropolitan aereas of Europe connsisting of 21 districts subdivided in 131 neighborhoods.
Madrid is multicultural city reflected in international cuisine restaurants opening constantly. In this project, I want to determine which is the best district and more specifically, which is the best neighborhood to open a new restaurant. Depending on the results, I will suggest to stakeholders which type of cuisine suits best for several locations.

### Data

I will be using Wikipedias page: List of neighborhoods of Madrid:
https://en.wikipedia.org/wiki/List_of_neighborhoods_of_Madrid 
The table will be imported into a pandas dataframe. 

Then, i will use Python geocoder library to get the latitude and longitude of each neighborhood. Both columns will be inserted into the dataframe.

Foursquare API will help in determining the most popular venues of each neighboorhood and also the ranks and likes of the restaurants that are on the top 3 popular venue list.
